```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from datetime import datetime
```

```python
def generar_gantt(tareas, titulo="Diagrama de Gantt"):
    """
    Genera un diagrama de Gantt a partir de una lista de tareas.

    Parámetros:
    - tareas: lista de diccionarios con claves: 'nombre', 'inicio', 'fin'.
              Ejemplo: [
                  {"nombre": "Fase 1", "inicio": "2024-06-01", "fin": "2024-06-10"},
                  {"nombre": "Fase 2", "inicio": "2024-06-11", "fin": "2024-06-20"},
              ]
    - titulo: título del gráfico
    """
    df = pd.DataFrame(tareas)
    df["inicio"] = pd.to_datetime(df["inicio"])
    df["fin"] = pd.to_datetime(df["fin"])
    df["duracion"] = df["fin"] - df["inicio"]

    fig, ax = plt.subplots(figsize=(10, 6))
    for i, tarea in df.iterrows():
        ax.barh(tarea["nombre"], tarea["duracion"].days,
                left=tarea["inicio"], height=0.5, color="skyblue", edgecolor="black")

    ax.set_xlabel("Fecha")
    ax.set_ylabel("Tareas")
    ax.set_title(titulo)
    ax.xaxis_date()
    ax.xaxis.set_major_formatter(mdates.DateFormatter("%d-%b"))

    plt.tight_layout()
    plt.grid(True, axis='x', linestyle='--', alpha=0.5)
    plt.show()
```
